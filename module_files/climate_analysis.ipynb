# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# reflect an existing database into a new model
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect the tables
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# finding a year back from the date specified
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)
```

```python
# performing the query to get date and prcp of all rows i think
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
```

```python
# very big result and hard to read all in one line list as tuple - ish
print(results)
```

```python
# setting the date column to index and sorting it
df = pd.DataFrame(results, columns=['date', 'percipitation']).set_index('date').sort_index()
df
```

```python
# plot the percip df
df.plot(title='Percipitation from 2016-2018', rot=90, xlabel='', figsize=(10,4))
```

```python
df.describe()
```

```python
# finding the count of stations
session.query(func.count(Station.station)).all()
```

```python
# count the entries in each station category
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station)\
.order_by(func.count(Measurement.station).desc()).all()
```

```python
# min max and average temp in the whole dataset
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs),\
              func.avg(Measurement.tobs)).filter(Measurement.station == 'USC00519281').all()
```

```python
# adding temp data in the last year for one year into results and then to df
results = session.query(Measurement.tobs).filter(Measurement.station == 'USC00519281')\
    .filter(Measurement.date >= prev_year).all()
df = pd.DataFrame(results, columns=['tobs'])
```
